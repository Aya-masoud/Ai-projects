```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("breast.csv")
```

```python
print(len(data.id.unique()))
```

```python
data.drop("id",inplace=True,axis=1)
```

```python
data.head(30)
```

```python
data.isna().sum()
```

```python
# data.diagnosis=data.diagnosis.replace({"M":1,"B":0})
data["diagnosis"]=data["diagnosis"].apply(lambda x: 1 if x=='M' else 0 )
```

```python
data.head()
```

```python
data.describe().T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr=data.corr()
```

```python
plt.figure(figsize = (30, 20))
mask = np.triu(np.ones_like(corr, dtype = bool))

sns.heatmap(data=corr,annot=True,mask=mask,linewidths=1,
            cmap="Blues",fmt = ".2f")
plt.show()
```

```python
data.diagnosis.value_counts()
```

```python
class_names = ["Benign", "Malignant"]
```

```python
X = data.drop('diagnosis', axis = 1)
y = data['diagnosis']
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 2)
```

```python
X_cols=X_train.columns
X_cols
```

```python

from sklearn.preprocessing import StandardScaler
sclar=StandardScaler()
X_train=sclar.fit_transform(X_train)
X_test=sclar.transform(X_test)
```

```python
X_train=pd.DataFrame(X_train,columns=X_cols)
X_test=pd.DataFrame(X_test,columns=X_cols)
```

```python
X_train.head()
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
svc=SVC()
param_grids={
            "kernel": ['linear', 'poly', 'rbf', 'sigmoid'],
             "C": [0.01, 0.05, 0.5, 0.1, 1, 10, 15, 20],
             "gamma" : [0.0001, 0.001, 0.01, 0.1]
}
svcgridModel=GridSearchCV(estimator=svc,param_grid=param_grids,cv=5,scoring="accuracy")
svcgridModel.fit(X_train,y_train)
```

```python
print(f" best_params_ : {svcgridModel.best_params_}")
best_est_model_SCV=svcgridModel.best_estimator_
```

```python
y_pred_SVC=best_est_model_SCV.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,auc
print(f"accuracy_score {accuracy_score(y_test,y_pred_SVC)}")
print(f"classification_report {classification_report(y_test,y_pred_SVC,target_names=class_names,digits=4)}")


cm_scv=confusion_matrix(y_test,y_pred_SVC)
plt.figure(figsize=(7,7))

sns.heatmap(data=cm_scv,annot=True,xticklabels=class_names,yticklabels=class_names,cmap="Blues",fmt="d")
plt.title('', size=24) 
plt.xlabel('Predicted label', size=12) 
plt.ylabel('True label', size=12) 
plt.show()

fpr,tpr,thersold=roc_curve(y_test,y_pred_SVC)
roc_auc=auc(fpr,tpr)
plt.figure(figsize=(6,7))
plt.plot(fpr,tpr,label='ROC curve (area = {:.4f})'.format(roc_auc),color='darkorange', lw=2)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('SVC  (ROC) Curve for breast cancer')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier


dtc = DecisionTreeClassifier()

parameters = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : range(2, 32, 5),
    'min_samples_leaf' : range(1, 10, 3),
    'min_samples_split' : range(2, 10, 3),
    'splitter' : ['best', 'random']
}

grid_search_dt = GridSearchCV(dtc, parameters, cv = 5, n_jobs = -1, verbose = 1)
grid_search_dt.fit(X_train, y_train)
print(grid_search_dt.best_params_)
```

```python
best_est_dtc=grid_search_dt.best_estimator_
y_pred_dtc=best_est_dtc.predict(X_test)
```

```python
print(f"accuracy_score decision tree {accuracy_score(y_test,y_pred_dtc)}")
print(f"classification_report {classification_report(y_test,y_pred_dtc,target_names=class_names,digits=4)}")


cm_scv=confusion_matrix(y_test,y_pred_dtc)
plt.figure(figsize=(5,5))

sns.heatmap(data=cm_scv,annot=True,xticklabels=class_names,yticklabels=class_names,cmap="Blues",fmt="d")
plt.title('', size=24) 
plt.xlabel('Predicted label', size=12) 
plt.ylabel('True label', size=12) 
plt.show()
```

```python
fpr,tpr,thersold=roc_curve(y_test,y_pred_dtc)
roc_auc=auc(fpr,tpr)
plt.figure(figsize=(5,5))
plt.plot(fpr,tpr,label='ROC curve (area = {:.4f})'.format(roc_auc),color='darkorange', lw=2)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(' decision tree (ROC) Curve for breast cancer')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier()
grid_search_RFC_param={
             "n_estimators":[100,250,300],
             "criterion":['gini', 'entropy'],
             "max_features": ['sqrt', 'log2'],
             'max_depth': [10, 11, 12]
}
grid_search_RFC=GridSearchCV(estimator=RFC,param_grid=grid_search_RFC_param,cv=5,scoring="accuracy",error_score=0)
grid_search_RFC.fit(X_train,y_train)
print(f" Random Forest Classifier best_params_{grid_search_RFC.best_params_}")
```

```python

best_est_RFC=grid_search_RFC.best_estimator_
y_prd_RFC=best_est_RFC.predict(X_test)

```

```python
print(f"accuracy_score Random Forest Classifier {accuracy_score(y_test,y_prd_RFC)}")
print(f"classification_report {classification_report(y_test,y_prd_RFC,target_names=class_names,digits=4)}")


cm_scv=confusion_matrix(y_test,y_prd_RFC)
plt.figure(figsize=(5,5))

sns.heatmap(data=cm_scv,annot=True,xticklabels=class_names,yticklabels=class_names,cmap="Blues",fmt="d")
plt.title('', size=24) 
plt.xlabel('Predicted label', size=12) 
plt.ylabel('True label', size=12) 
plt.show()
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prd_RFC)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random forest  (ROC) Curve for breast cancer')
plt.legend(loc='lower right')
plt.show()
```
