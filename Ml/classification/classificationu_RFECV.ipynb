```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv("dataset.csv",sep=";")
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(data.Target, palette="bwr")
plt.show()
data.Target.value_counts()
```

```python
cols_name=data.columns
```

```python
cols_name
```

```python
from category_encoders import OrdinalEncoder 
encoder=OrdinalEncoder(cols=["Target"])
data=encoder.fit_transform(data)
data=pd.DataFrame(data,columns=cols_name)
data.head()
```

```python
data.Target.value_counts()
```

```python
data = data[data['Target'] != 3]
```

```python
X = data.drop("Target",axis=1)
y = data["Target"]
```

```python
from sklearn import preprocessing
X= preprocessing.StandardScaler().fit_transform(X)
```

```python
X=pd.DataFrame(X,columns=data.drop("Target",axis=1).columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
model=RFECV(estimator=rfc,cv=5,step=22)
model.fit(X_train,y_train)
y_pred_rfc=model.predict(X_test)

```

```python
print(f'Optimal number of features: {model.n_features_}')
print(f'Selected features: {list(X.columns[model.support_])}')
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,auc
print("RandomForestClassifier")
print(classification_report(y_test, y_pred_rfc,digits=4))

cm = confusion_matrix(y_test, y_pred_rfc)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

#  Draw the roc curev and auc using the results of random forest
# ROC Curve and AUC
y_pred_proba = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba,pos_label=2)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression,SGDClassifier
lr=LogisticRegression(max_iter=1000)
model=RFECV(estimator=lr,step=25,cv=5)
model.fit(X_train,y_train)
y_pred_lr=model.predict(X_test)


print(f'Optimal number of features: {model.n_features_}')
print(f'Selected features: {list(X.columns[model.support_])}')
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,auc
print("LogisticRegression")
print(classification_report(y_test, y_pred_lr,digits=4))

cm = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()


y_pred_proba = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba,pos_label=2)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression,SGDClassifier
sGDClassifier=SGDClassifier(max_iter=1000,loss="log")
model=RFECV(estimator=sGDClassifier,step=25,cv=5)
model.fit(X_train,y_train)
y_pred_lr=model.predict(X_test)



print(f'Optimal number of features: {model.n_features_}')
print(f'Selected features: {list(X.columns[model.support_])}')
```

```python
from sklearn.model_selection import GridSearchCV
sGDClassifier=SGDClassifier()
para_grid={
             "loss": ['hinge', 'log_loss', 'log', 'modified_huber', 'squared_hinge', 'perceptron'],
              "penalty": ['l2', 'l1', 'elasticnet'] ,
              "alpha": [ 0.0001,0.001,0.01,0.1],
              "max_iter": [1000,1100,1300],
            
}
grid_model=GridSearchCV(estimator=sGDClassifier,scoring="accuracy",param_grid=para_grid,cv=5,verbose=-1,n_jobs=-1)

grid_model.fit(X_train, y_train)

model=RFECV(estimator=grid_model.best_estimator_,step=25,cv=5)
model.fit(X_train,y_train)
y_pred_lr=model.predict(X_test)
```
