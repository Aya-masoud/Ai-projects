```python
X = data.drop("Target",axis=1)
y = data["Target"]
```

```python
X_cols=X.columns
X_cols
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaler=scaler.fit_transform(X)
X_scaler=pd.DataFrame(X_scaler,columns=X_cols)
X_scaler.head()
```

```python
from sklearn.feature_selection import SelectPercentile,chi2
selkbest=SelectPercentile(score_func=chi2,percentile=70)
new_X=selkbest.fit_transform(X_scaler,y)
new_X.shape 
```

```python
selkbest.get_support()
```

```python
print(X.columns[selkbest.get_support()].to_list())
```

```python
X_cols_select_y=X.columns[selkbest.get_support()].to_list()
X_cols_select_y.append("target")
X_cols_select_y=pd.DataFrame(X_cols_select_y)

print(X_cols_select_y)
X_cols_select=X.columns[selkbest.get_support()].to_list()

X_cols_select=pd.DataFrame(X_cols_select)
```

```python
X=new_X
```

```python
data=np.column_stack([X,y])
data=pd.DataFrame(data,columns=X_cols_select_y)
plt.figure(figsize=(30,30))
sns.heatmap(data.corr(),annot=True,fmt="0.3f",linewidths=0.2)
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=True)
X_train.shape ,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
dtc=DecisionTreeClassifier()
param_grid_dtc={
            "criterion": ['gini', 'entropy'],
             "splitter": ['best', 'random'],
             "max_depth":np.arange(1,12,1),
             "min_samples_split":np.arange(2,5,1),
             "min_samples_leaf":np.arange(1,5,1)
}
Grid_modelDtc=GridSearchCV(estimator=dtc,cv=5,error_score=0,param_grid=param_grid_dtc,scoring="accuracy",n_jobs=3,verbose=-1)
Grid_modelDtc.fit(X_train,y_train)
best_est_dtc=Grid_modelDtc.best_estimator_
y_pred_dtc=best_est_dtc.predict(X_test)
print(f" best_params_ {Grid_modelDtc.best_params_}")
print(f" best_score_ {Grid_modelDtc.best_score_}")
```

```python
from sklearn.metrics import classification_report,accuracy_score,auc,confusion_matrix,roc_curve
accuracy_dtc=accuracy_score(y_test,y_pred_dtc)
print(f"accuracy_score {accuracy_dtc}")
print(f"classification_report {classification_report(y_test,y_pred_dtc)}")
cm=confusion_matrix(y_test,y_pred_dtc)
sns.heatmap(cm,cmap="Blues",annot=True,fmt="d")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
y_pred_dtc_pro=best_est_dtc.predict_proba(X_test)[:,1]
fpr,tpr,thershold=roc_curve(y_test,y_pred_dtc_pro)
aucdtc=auc(fpr,tpr)
plt.plot(fpr,tpr,color='darkorange',lw=2,label='ROC curve (area = %0.4f)' % aucdtc)
plt.plot([0,1],[0,1],lw=2,color='navy',linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=42)
param_grid_rfc={
            "n_estimators":[100,200,250],
            "criterion": ['gini', 'entropy' ],
        'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
Grid_model_rfc=GridSearchCV(estimator=rfc,cv=5,error_score=0,param_grid=param_grid_rfc,scoring="accuracy",n_jobs=-1,verbose=-1)
Grid_model_rfc.fit(X_train,y_train)
best_est_rfc=Grid_model_rfc.best_estimator_
y_pred_rfc=best_est_rfc.predict(X_test)
print(f" best_params_ {Grid_model_rfc.best_params_}")
print(f" best_score_ {Grid_model_rfc.best_score_}")
```

```python
accuracy_rfc=accuracy_score(y_test,y_pred_rfc)
print(f"accuracy_score RandomForestClassifier {accuracy_rfc}")
print(f"classification_report {classification_report(y_test,y_pred_rfc)}")
cm=confusion_matrix(y_test,y_pred_rfc)
sns.heatmap(cm,cmap="Blues",annot=True,fmt="d")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
y_pred_rfc_pro=best_est_rfc.predict_proba(X_test)[:,1]
fpr,tpr,thershold=roc_curve(y_test,y_pred_rfc_pro)
aucrfc=auc(fpr,tpr)
plt.plot(fpr,tpr,color='darkorange',lw=2,label='ROC curve (area = %0.4f)' % aucrfc)
plt.plot([0,1],[0,1],lw=2,color='navy',linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV,StratifiedKFold
from sklearn.metrics import roc_auc_score
xgb=XGBClassifier(learning_rate=0.2,n_estimator=400, objective='binary:logistic',silent=True, nthread=1)
params_xgb = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5]
        }
stk=StratifiedKFold(n_splits=5,random_state=42,shuffle=True)
cv=stk.split(X_train,y_train)
gridrandom_Xgb=RandomizedSearchCV(estimator=xgb,n_iter=5,param_distributions=params_xgb,
                                  scoring="accuracy", n_jobs=-1,verbose=3,random_state=1001)
gridrandom_Xgb.fit(X_train,y_train)
best_XGB=gridrandom_Xgb.best_estimator_
y_pred_xgb = best_XGB.predict(X_test)
print(gridrandom_Xgb.best_params_, gridrandom_Xgb.best_score_)

```

```python
accuracy_xgb=accuracy_score(y_test,y_pred_xgb)
print(f"accuracy_score {accuracy_xgb}")
print(f"classification_report {classification_report(y_test,y_pred_xgb)}")
cm=confusion_matrix(y_test,y_pred_xgb)
sns.heatmap(cm,cmap="Blues",annot=True,fmt="d")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
y_pred_xgb_pro=best_XGB.predict_proba(X_test)[:,1]
fpr,tpr,thershold=roc_curve(y_test,y_pred_xgb_pro)
aucxgb=auc(fpr,tpr)
plt.plot(fpr,tpr,color='darkorange',lw=2,label='ROC curve (area = %0.4f)' % aucxgb)
plt.plot([0,1],[0,1],lw=2,color='navy',linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.svm import SVC
svc=SVC(probability=True)
param_grid_svc={
            "C": np.arange(0.5,1,0.1),
            "kernel": ['linear', 'poly', 'rbf', 'sigmoid'],
             "gamma":np.arange(0.01,0.05,0.01),
             "max_iter":[100,150,200,300]

}
STKF=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
cv2=STKF.split(X_train,y_train)
grid_search_svc=GridSearchCV(estimator=svc,cv=cv2,param_grid=param_grid_svc,n_jobs=-1,scoring="accuracy",error_score=0,verbose=-1)
grid_search_svc.fit(X_train,y_train)
best_SVC=grid_search_svc.best_estimator_
y_pred_SVC = best_SVC.predict(X_test)
print(grid_search_svc.best_params_, grid_search_svc.best_score_)
```

```python
accuracy_SVC=accuracy_score(y_test,y_pred_SVC)
print(f"accuracy_score SVC {accuracy_SVC}")
print(f"classification_report \n {classification_report(y_test,y_pred_SVC,digits=4)}")
cm=confusion_matrix(y_test,y_pred_SVC)
sns.heatmap(cm,cmap="Blues",annot=True,fmt="d")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
y_pred_SVC_pro=best_SVC.predict_proba(X_test)[:,1]
fpr,tpr,thershold=roc_curve(y_test,y_pred_SVC_pro)
aucSCV=auc(fpr,tpr)
plt.plot(fpr,tpr,color='darkorange',lw=2,label='ROC curve (area = %0.4f)' % aucSCV)
plt.plot([0,1],[0,1],lw=2,color='navy',linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.ensemble import VotingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.neural_network import MLPClassifier
vcl=VotingClassifier(estimators=[

 ('svm', SVC(probability=True)),
    ('lr', LogisticRegression()),
    ('rf', RandomForestClassifier(n_estimators= 300)),
    ('dt', DecisionTreeClassifier(max_depth=6, min_samples_leaf=1, min_samples_split=2, splitter= 'random')),
    ('xgb', XGBClassifier(subsample= 0.8, min_child_weight= 5, max_depth= 5, gamma= 1, colsample_bytree= 0.8)),
    ('mlp', MLPClassifier(activation= 'relu', alpha= 0.0001, hidden_layer_sizes= (20,), learning_rate= 'adaptive', solver= 'adam')),
    ('bagging',BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=6, min_samples_leaf=1, min_samples_split=2, splitter= 'random'))),
    ('adaboost',AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=6, min_samples_leaf=1, min_samples_split=2, splitter= 'random')))
    ], voting='soft')
params = {'lr__C': [1.0, 100.0],
      'svm__C': [2,3,4],
          }
grid = GridSearchCV(estimator=vcl, param_grid=params, cv=5,n_jobs=-1, verbose=3)

grid.fit(X_train,y_train)
print (grid.best_params_)
best_VC=grid.best_estimator_
y_pred=best_VC.predict(X_test)

```

```python

print(classification_report(y_test, y_pred,digits=4))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()


y_pred_proba = best_VC.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import StackingClassifier
level0 = list()
level0.append(('LR', LogisticRegression(C= 0.1, penalty='l2')))
level0.append(('RF',RandomForestClassifier(n_estimators= 300)))
level0.append(('DT', DecisionTreeClassifier(max_depth=6, min_samples_leaf=1, min_samples_split=2, splitter= 'random')))
level0.append(('SGD',SGDClassifier(alpha= 0.01)))
level0.append(('XGB',XGBClassifier(subsample= 0.8, min_child_weight= 5, max_depth= 5, gamma= 1, colsample_bytree= 0.8)))
level0.append(('MLP',MLPClassifier(activation= 'relu', alpha= 0.0001, hidden_layer_sizes= (20,), learning_rate= 'adaptive', solver= 'adam')))
level0.append(('Adaboost',AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=6, min_samples_leaf=1, min_samples_split=2, splitter= 'random'))))
level0.append(('bagging',BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=6, min_samples_leaf=1, min_samples_split=2, splitter= 'random'))))
```

```python
level1= SVC(C= 0.7999999999999999, gamma= 0.04, kernel= 'rbf')

model = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred,digits=4))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
